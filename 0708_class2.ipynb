```python
#11866 요세푸스 문제 0
# K 번째 사람 제거  / N 명 사람 모두 제거될 때까지 계속함
from collections import deque
N,K = map(int, input().split())

queue = deque()
for i in range(1,N+1):
    queue.append(i)

index = K -1 # index는 0부터 시작하기 때문
print('<', end='')

while len(queue)>1:
    for i in range(K-1):
        queue.append(queue.popleft()) #K-1까지 원소 꺼내서 뒤에 다시 삽입
    print(queue.popleft(), end = ', ')

print(queue.popleft(),end='')
print('>')
```

```python
#10866 덱
from collections import deque
import sys
t = int(input())
queue = deque()
while t>0:
    temp = sys.stdin.readline().split()
    if temp[0] == 'push_back':
        queue.append(int(temp[1]))
        ""
    elif temp[0] == 'push_front':
        queue.appendleft(int(temp[1]))
        ""
    elif temp[0] == 'pop_front':
        if queue:
            print(queue[0])
            queue.popleft()
        else:
            print(-1)
    elif temp[0] == 'pop_back':
        if queue:
            print(queue[-1])
            queue.pop()
        else:
            print(-1)
        ""
    elif temp[0] == 'size':
        print(len(queue))
    elif temp[0] == 'empty':
        if len(queue) == 0:
            print(1)
        else:
            print(0)
    elif temp[0] == 'front':
        if queue:
            print(queue[0])
        else:
            print(-1)
    
    else:
        if queue:
            print(queue[-1])
        else:
            print(-1)
    t-=1
    
    
```
